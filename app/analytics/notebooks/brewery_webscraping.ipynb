# Imports

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import csv
```

# Get brewerys 

```python
cervecerias = []
cervecerias_not_null = []
cervecerias_matching = []
for i in range(41):
    page = requests.get('https://birrapedia.com/cervecerias/de-espana?par_pagina={}&cl_tipo=ce'.format(i))
    print(i, page)
    soup = BeautifulSoup(page.content, 'html.parser')
    #sacamos el valor de cada href
    for data in soup.find_all('div', class_='row'):
        for a in data.find_all('a'):
            cervecerias.append(a.get('href'))
    #borramos None de la lista
    for i in cervecerias:  
        if i:
            cervecerias_not_null.append(i)
    #añadimos los elementos que nos interesan a la lista
    for href in cervecerias_not_null:
        if "/cervecerias/" in href:
            cervecerias_matching.append(href)
```

```python
aux = set(cervecerias_matching)
cervecerias_matching= list(aux)
```

## Save brewerys

```python
with open('/mnt/c/Users/cx02379/Documents/cervesanal/cervecerias.csv', 'w', newline='') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(cervecerias_matching)
```

# Get beers

```python
cervezas = []
cervezas_matching = []
for i in cervecerias_matching:
    #print("cerveceria: ", i)
    page = requests.get('https://birrapedia.com{}'.format(i))
    soup = BeautifulSoup(page.content, 'html.parser')
    #sacamos el valor de cada href
    for data in soup.find_all('div', class_='col-xs-4'):
        for a in data.find_all('a'):
            cervezas.append(a.get('href'))
    #añadimos los elementos que nos interesan a la lista
    for href in cervezas:
        if  '/cervezas/' in href:
            cervezas_matching.append(href)
```

```python
aux = set(cervezas_matching)
cervezas_matching= list(aux)
```

# Get info

```python
len(cervezas_matching)
```

```python
df_total = pd.DataFrame(columns=('Name', 'Brewery', 'Style', 'ABV', 'IBU', 'Rating', 'Description')) 

for cerveza in cervezas_matching:
    print(cerveza)
    print(cervezas_matching.index(cerveza))
    page = requests.get('https://birrapedia.com{}'.format(cerveza))
    soup = BeautifulSoup(page.content, 'html.parser')
    #sacamos la descripcion
    for data in soup.find_all(class_='text-muted mt-1'):
        description = data.get_text()
    #sacamos el nombre
    for data in soup.find_all("h1", class_='mdl-typography--title-color-contrast m-0 p-0'):
        name = data.get_text()
    #sacamos el estilo:
    for data in soup.find_all(class_='mr-1'):
        for a in data.find_all('a'):
            style = data.get_text()
    #alcohol e ibu
    for data in soup.find_all(class_='mr-1'):
        abv = data.get_text()
        
    for value, data in enumerate(soup.find_all(class_='mt-1')):
        if value == 4:
            brewery = data.get_text()
    df = pd.DataFrame({"Name": name, 
                       "Brewery": brewery,
                       "Style": style,
                       "ABV": abv, 
                       'IBU': abv, 
                       'URL': cerveza,
                       'Rating': None, 
                       'Description': description}, index=[0]) 
    
    df_total = df_total.append(df)
```
